```python
import gensim
import wikipedia
import nltk
from tqdm import tqdm
```

```python
mona_lisa=wikipedia.page('Mona Lisa')
eifel=wikipedia.page('Eifel Tower')
train=wikipedia.page('Trains')
data=wikipedia.page('Data Science')
```

example of corpus obtained

```python
mona_lisa.content
```

```python
corpus=[mona_lisa.content,eifel.content,train.content,data.content]
```

# preprocessing the text

```python
import re
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
en_stop=nltk.corpus.stopwords.words('english')

stemmer=WordNetLemmatizer()

def preprocess_text(doc):
    #remove special chars
    doc=re.sub(r'\W',' ',doc)
    
    #remove single chars
    doc=re.sub(r'\s+[a-zA-Z]\s+',' ',doc)
    
    #remove single chars from start
    doc=re.sub(r'\s+[a-zA-Z]\s+',' ',doc)
    
    #remove multiple spaces with single space
    doc=re.sub(r'\s+',' ',doc,flags=re.I)
    
    #remove prefixed 'b'
    doc=re.sub(r'^b\s+','',doc)
    
    doc=doc.lower()
    tokens=doc.split()
    tokens=[stemmer.lemmatize(word) for word in tokens]
    tokens=[word for word in tokens if word not in en_stop]
    tokens=[word for word in tokens if len(word) > 5]
    
    return tokens
```

```python
clean_corpus=[]
for docs in tqdm(corpus):
    cleaned_tokens=preprocess_text(docs)
    clean_corpus.append(cleaned_tokens)
```

# LDA Model

```python
from gensim import corpora
```

```python
input_dict=corpora.Dictionary(clean_corpus)
input_corpus=[input_dict.doc2bow(token,allow_update=True) for token in clean_corpus]
```

```python
input_corpus[:6]
```

```python
input_dict.get(2473)
```

```python
lda_model=gensim.models.ldamodel.LdaModel(input_corpus,num_topics=4,id2word=input_dict,passes=20)
```

```python
topics=lda_model.print_topics(num_words=10)
topics
```

```python
test_text='machine learning uses data visualization,cleaning and preprocessing followed by training and testing of mathematical models.'
```

the test text should belong to topic 3 with high probab

```python
test=preprocess_text(test_text)
test=input_dict.doc2bow(test)

res_probabs=lda_model.get_document_topics(test)
res_probabs
```
